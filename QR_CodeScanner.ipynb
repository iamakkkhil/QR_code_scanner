```
qr = cv2.imread("/content/QR_2.jpeg")
```

```
points = []
for info in decode(qr):
    data = info.data.decode()
    for pts in info.polygon:
        points.append([pts.x, pts.y])
print(data)
```

```
points = np.array(points, np.int32)
points = points.reshape((-1, 1, 2))
```

```
cv2.polylines(qr, [points], True, (255,0,0), 2)
cv2.putText(qr, data, (100,100), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1, cv2.LINE_AA )
cv2_imshow(qr)
```
